# Import Data

```python
import pandas as pd

df = pd.read_csv('Indoor Air Pollution Data.csv')
df
```

# Data Cleaning & Preprocessing

```python
# Drop unused columns

for i in range (9, 19):
  df.drop(f'Unnamed: {i}', axis=1, inplace=True)
df.head()
```

```python
df.info()
```

```python
df['NO2'].unique()
```

```python
weird_values = ['2021-04-06|07:29:56.494601', '2021-05-10|09:52:53.649941', '2021-05-10|09:53:03.434813']
df[df['NO2'].isin(weird_values)]
```

```python
# Drop weird record
df = df[~df['NO2'].isin(weird_values)]
df
```

```python
df[df['O3'].isin(['12.450.73'])]
```

```python
df = df[~df['O3'].isin(['12.450.73'])]
df
```

```python
# Type casting to float
df['NO2'] = df['NO2'].astype(float)
df['Temp'] = df['Temp'].astype(float)
df['Pressure'] = df['Pressure'].astype(float)
df['Humidity'] = df['Humidity'].astype(float)
df['O3'] = df['O3'].astype(float)
df
```

```python
df.info()
```

```python
from datetime import datetime

# Define the function to convert date strings to standard format
def convert_date_format(date_str):
    try:
        # Replace '|' with ' ' to match the standard format
        standard_date_str = date_str.replace('|', ' ')

        # Parse the date string to a datetime object
        parsed_date = datetime.strptime(standard_date_str, '%Y-%m-%d %H:%M:%S.%f')

        # Return the formatted date string in standard format
        return parsed_date.strftime('%Y-%m-%d %H:%M:%S.%f')
    except ValueError:
        # Handle the case where date_str is not in the expected format
        return None

# Apply the function to the 'Date' column
df['Date'] = df['Date'].apply(convert_date_format)

# Display the DataFrame with the converted dates
df
```
